```python
lista = []
contador = 0

for index, row in df.iterrows(): 
    country = row['Con']
    temp_row = row
    if (country.find("/")!=-1):
        #print(country)        
        for pais in country.split("/"):
            new_row = temp_row[1:]
            new_row["Con"] = pais
            #print('1 -->',pais)
            contador = contador + 1
            lista.append(new_row)
    else:
        #print('2 -->',temp_row["Con"])
        contador = contador + 1
        lista.append(temp_row)

print("Num reg:",contador)

```
